# 1. 데이터 탐색 (EDA)

건강 검진 데이터를 기반으로 흡연 여부를 예측하는 이진 분류 프로젝트입니다.  
이 노트북에서는 데이터를 처음 마주한 분석가의 시선으로 관찰하고, 결측치·이상치를 확인하며, 변수 간 상관관계를 분석합니다.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.rc('font', family='Apple SD Gothic Neo')
plt.rc('axes', unicode_minus=False)
```

## 1.1 데이터 로드 및 기본 정보 확인

```python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')

print(f'Train shape: {train.shape}')
print(f'Test shape: {test.shape}')
train.head()
```

```python
train.info()
```

```python
train.describe()
```

## 1.2 결측치 확인

```python
print("Train 결측치 수:\n", train.isnull().sum())
print(f"\nTrain 총 결측치: {train.isnull().sum().sum()}")
print(f"Test 총 결측치: {test.isnull().sum().sum()}")
```

## 1.3 타겟 변수 분포 확인

```python
sns.countplot(x='label', data=train, palette=['#4CAF50', '#FF7043'])
plt.title('Smoking Distribution')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Smoker', 'Smoker'])
plt.show()

print(f"클래스 비율:\n{train['label'].value_counts(normalize=True)}")
```

## 1.4 변수별 분포 및 이상치 확인 (히스토그램 + 박스플롯)

```python
numeric_cols = train.drop(['ID', 'label'], axis=1).columns

for col in numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    axes[0].hist(train[col], bins=30)
    axes[0].set_title(f"{col} Histogram")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("count")

    axes[1].boxplot(train[col], vert=False)
    axes[1].set_title(f"{col} Boxplot")
    axes[1].set_xlabel(col)
    plt.tight_layout()
    plt.show()
```

## 1.5 상관관계 분석

```python
numeric_df = train.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="RdBu_r", square=True)
plt.title("Numeric Variables Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## 1.6 타겟(label)과의 상관관계 Top 10

```python
corr_label = numeric_df.corr()['label'].drop('label')
top10 = corr_label.abs().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
plt.bar(x=top10.index, height=top10.values, color='gray')
plt.xticks(rotation=30)
plt.title("흡연 여부와 상관계수 Top 10 변수")
plt.ylabel("상관계수 (절대값)")
plt.tight_layout()
plt.show()
```
